```python
#Encoding categorical data values (Transforming object data types to integers)
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

#Encode sex column
df.iloc[:,2]= labelencoder.fit_transform(df.iloc[:,2].values)
#print(labelencoder.fit_transform(titanic.iloc[:,2].values))
print(df['sex'].unique())
```

```python

# Filter DataFrame for survived passengers
survived_df = df[df['survived'] == 1]

# Count the number of survived males and females
survived_counts = survived_df['sex'].value_counts()
print(survived_counts)

# Plot countplot for survived passengers with males as 0 and females as 1
sns.countplot(x='sex', data=survived_df, hue='survived')
plt.title('Count of Survived Passengers by Sex')
plt.xlabel('Sex (0 = Male, 1 = Female)')
plt.ylabel('Count')
plt.show()
```

```python
#Count the empty (NaN, NAN, na) values in each column
df.isna().sum()
```

```python
# Drop the columns
df = df.drop(['deck', 'embark_town', 'alive', 'class', 'alone', 'adult_male', 'who'], axis=1)

#Remove the rows with missing values
df = df.dropna(subset =['embarked'])
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

# Define X and Y
X = df[['sex', 'pclass']]
Y = df['survived']

# Split the dataset into 80% Training set and 20% Testing set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
log =LogisticRegression(random_state=0)
log.fit(X_train,Y_train)
```

```python
 print('[0]Logistic Regression Training Accuracy:', log.score(X_train, Y_train))
```

```python
pred=print(log.predict(X_test))
```
